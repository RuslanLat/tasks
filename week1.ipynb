### Задание 1
Создать бинарный файл, заполнив разными случайными данными: буквы, числа, знаки препинания. \
Далее считать данные из этого файла, вывести на экран, вывести статистику: число символов, слов, строк. \
В качестве ответа приложите ссылку на выполненное задание в вашем репозитории (на соотв. файл/ветку)

```python
# импорт необходимых библиотек
import string
import random
```

```python
# буквы, числа, знаки препинания
letters = string.ascii_letters
digits = string.digits
punctuation = string.punctuation
datas = letters + digits + punctuation
```

```python
# список случайных букв, чисел, знаков препинания
random.choices(datas, k=10)
```

```python
# создание бинарного файла, заполненного разными случайными данными (буквы, числа, знаки препинания)
with open("data/task1.bin", "wb") as f:
    lines = [line.encode() + b"\n" for line in random.choices(datas, k=200)]
    f.writelines(lines)
        
```

```python
# считывание бинарного файла
with open("data/task1.bin", "rb") as f:
    lines = f.readlines()
```

```python
# вывод на экран содержимого файла
for line in lines:
    print(line)
```

```python
# подсчет статистики
count_letters = 0
count_digits = 0
count_punctuation = 0
for line in lines:
    if line.decode().strip() in letters:
        count_letters += 1
    elif line.decode().strip() in digits:
        count_digits += 1
    else:
        count_punctuation += 1
print(f"В файле {count_letters} букв")
print(f"В файле {count_digits} чисел")
print(f"В файле {count_punctuation} знаков препинаний")
```

### Задание 2
Вывести структуру(в виде дерева) указанной папки, переданной как аргумент, используя модуль os, отображать имя и размер каждого файла.

```python
# импорт необходимых библиотек
import os
from os.path import join, getsize, basename
```

```python
# текущая директория
directory = os.getcwd()
directory
```

```python
def dfs_tree(path):
    def tree(path, level):
        dsize, files, dirs = 0, [], []
        # формирование структуры папки (в виде дерева)
        files.append(f'{" " * (level - 1) * 4}\\{basename(path)}\\ ')
        # перебор в цикле всех файлов и папок в отсортированном списке (сначала папки, потом файлы)
        for item in sorted(
            os.listdir(path), key=lambda x: os.path.isfile(join(path, x))
        ):
            # путь к файлу или папке
            fullname = join(path, item)
            # условие если это папка
            if os.path.isdir(fullname):
                # рекурсивный вызов функции
                tmp = tree(fullname, level + 1)
                # подсчет размера папки
                dsize += tmp[0]
                # формирование структуры папки
                files.append(tmp[1])
            # условие если это файл
            else:
                # размер файла
                fsz = getsize(fullname)
                # формирование структуры папки (в виде дерева)
                files.append(f'{" " * level * 4}{item}  [{fsz:,} bytes]')
                # подсчет размера папки
                dsize += fsz
        # формирование структуры папки (в виде дерева)       
        files[0] += f"[{dsize:,} bytes]"

        return dsize, "\n".join(files)

    return tree(path, 1)[1]
```

```python
# вывод структуры в виде дерева указанной папки
print(dfs_tree(directory))
```

### Задание 4
С помощью функции reduce и лямбда-функции объединить список строк в одну строку, разделяя части пробелами, \
преобразуя в нижний регистр. \
Пример: [‘Privet’, “mir”] -> ‘privet mir’

С помощью функции filter и лямбда-функции реализовать фильтр последовательности с логикой по своему усмотрению, \
логику описать в комментарии.

С помощью функции map сделать отображение одной последовательности (списка, множества или строки), \
состоящий из эл-тов РАЗНЫХ типов на другую по своей логике (описать логику).

```python
# импорт необходимых библиотек
from functools import reduce
```

```python
# список строк
text_list = ["Привет", "МИР", "ЗаданиE 4"]
```

```python
# объединение списка строк в одну строку, части разделены пробелами
reduce(lambda x, y: x.lower() + " " + y.lower(), text_list)
```

```python
# список чисел
numbers = [1, 2, 4, 5, 7, 8, 10, 11]
```

```python
# фильтрация четных чисел
list(filter(lambda x: x%2 == 0, numbers))
```

```python
# список с разными типами данных
data = [1, 33, -5, -0.1111, "33", 8]
```

```python
# приведение списка к одному типу данных
list(map(int, data))
```
